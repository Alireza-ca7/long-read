```python
import pywt
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample signal
signal = np.sin(2 * np.pi * 3 * np.linspace(0, 1, 1024)) + np.sin(2 * np.pi * 10 * np.linspace(0, 1, 1024))

# Perform Discrete Wavelet Transform (DWT) using Daubechies wavelet
wavelet = 'db1'  # Daubechies wavelet with 1 vanishing moment (Haar wavelet)
coeffs = pywt.wavedec(signal, wavelet)

# Plot original signal and wavelet coefficients
plt.figure(figsize=(12, 20))

plt.subplot(2, 1, 1)
plt.plot(signal)
plt.title('Original Signal')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

for i in range(len(coeffs)):
    plt.subplot(2, len(coeffs), len(coeffs) + i + 1)
    plt.plot(coeffs[i])
    plt.title(f'Level {i+1} Coefficients')
    plt.xlabel('Coefficient Index')
    plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

```python
def normalize_read_counts(read_counts):
    """
    Normalize read counts to correct for differences in library size.

    Args:
        read_counts: List containing the read counts per genomic bin.

    Returns:
        normalized_counts: List of normalized read counts.
    """
    total_reads = sum(read_counts)
    normalized_counts = [count / total_reads for count in read_counts]
    return normalized_counts

def smooth_read_counts(read_counts, window_size=5):
    """
    Smooth read counts using a moving average.

    Args:
        read_counts: List containing the read counts per genomic bin.
        window_size: Size of the moving average window.

    Returns:
        smoothed_counts: List of smoothed read counts.
    """
    kernel = np.ones(window_size) / window_size
    smoothed_counts = np.convolve(read_counts, kernel, mode='same')
    return smoothed_counts
```

```python
import pysam
import numpy as np
import pywt

def detect_cnvs(bam_file, window_size=1000, min_amplitude=0.5):
    """
    Detect copy number variations (CNVs) from a BAM file using wavelet transform.

    Args:
        bam_file (str): Path to the BAM file.
        window_size (int): Size of the sliding window in base pairs.
        min_amplitude (float): Minimum amplitude threshold for CNV detection.

    Returns:
        list: List of detected CNVs in the format [(chromosome, start, end, copy_number), ...]
    """
    # Open the BAM file
    bam = pysam.AlignmentFile(bam_file, "rb")

    # Initialize the CNV list
    cnvs = []

    # Iterate through the chromosomes
    for chrom in bam.references:
        # Get the read depths for the chromosome
        depths = np.zeros(bam.get_reference_length(chrom), dtype=int)
        for read in bam.fetch(chrom):
            depths[read.reference_start:read.reference_end] += 1

        # Apply the wavelet transform
        wavelet, _ = pywt.cwt(depths, np.arange(1, 11), 'gaus1')

        # Detect CNVs
        for i in range(0, len(depths) - window_size, window_size):
            window = depths[i:i + window_size]
            wavelet_amp = np.max(np.abs(wavelet[:, i:i + window_size]), axis=0)
            if np.max(wavelet_amp) > min_amplitude:
                start = i
                end = i + window_size
                copy_number = np.mean(window) / np.mean(depths)
                cnvs.append((chrom, start, end, copy_number))

    # Close the BAM file
    bam.close()

    return cnvs
```
